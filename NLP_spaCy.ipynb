```python
import spacy
nlp = spacy.load('en_core_web_sm') # small english model for testing some tokenization tecniques
```

```python
with open ("/content/sample_data/germany_article.txt", "r") as f:
  text = f.read()
```

```python
print(text)
```

```python
doc = nlp(text)
```

```python
print(doc)
```

```python
print(len(text))
print(len(doc))
```

```python
for token in text[:10]:
  print(token)
```

```python
for token in doc[:10]:  # tokenization is powerfull in NLP
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```

```python
sentence1 = list(doc.sents)[1]  # we need to pass the doc.sents in a list so that we can access its Content
print(sentence1)
```

```python
doc2 = nlp(text)

for token in doc2:
  print(token.text, token.pos_, token.dep_)
```

### Importing Displacy

```python
from spacy import displacy
displacy.render(doc2, style="dep")
```

```python
displacy.render(doc2, style="ent") # entities displayed with colors
```
